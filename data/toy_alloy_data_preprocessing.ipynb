```python
#Calculate value counts 
def cal_value_counts(df, col_index, threthold):
    df_col = df.iloc[:, col_index].value_counts()
    df_sel = df_col[df_col > threthold]
    df_unique = df_col[df_col == 1]
    unique_counts = df_unique.shape[0]
    unique_counts_pct = np.round(unique_counts/df.shape[0]*100,2)
    print ('Number of unique value for ' + df_unique.name + ':', unique_counts, '('+ str(unique_counts_pct) +'%)')
    return df_sel
```

## 2.Feature engineering

```python
X, Y = df_1.iloc[:,:-6], df_1.iloc[:, -6:]
```

```python
# mixing entropy: S/R = sum of X*In(X)
S = X*np.log(X)
S.fillna(0, inplace=True)
s = S.sum(axis = 1).to_frame(name = 'mixing entropy')
```

```python
# mixing enthalpy: 4* sum of XiXj*Hij
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(include_bias=False, interaction_only =True)
H = poly.fit_transform(X)
new_col_names = poly.get_feature_names(X.columns.tolist())[12:]
```

```python
h = pd.DataFrame(data = H[:, 12:], columns = new_col_names, index= X.index)
```

```python
h
```

```python
# PHS = H*S/R
phs = h.multiply(s.values, axis=0).add_prefix('phs_')
```

```python
df_3_1 = pd.concat([X, s, h, Y], axis=1)
```

```python
df_3_1.to_csv('preprocessed/data_fe_1.csv', index=False)
```

## With missing data

```python
df_missing = pd.read_csv('toy_alloy_data_with_missings.csv')
```

```python
df_missing.head()
```

```python
df_missing.isnull().sum(axis=1)
```

```python
for i in range(8):
    if i == 6:
        num = cal_duplication(df_missing, list(range(0,12))).shape[0]
        print ('number of duplications in all elements', num )
    elif i ==7:
        num = cal_duplication(df_missing, list(range(0,18))).shape[0]
        print ('number of duplications in all elements and all properties', num )
    else:
        num = cal_duplication(df_missing, list(range(0,12)) + [i+12]).shape[0]
        print('number of duplications in all elements and property ' + str(i+1), num)
```

```python
df_elements, df_properties =  df_missing.iloc[:,:12], df_missing.iloc[:,12:]
df_null_sum = df_elements.isnull().sum(axis=1)
```

```python
# if there is only one missed value, this value can be calculated according to the rule that the summation of all elements is one
sum_elements = df_elements[df_null_sum==1].sum(axis=1)
fill_values = 1- sum_elements
```

```python
#fill missing value when there is only one missing value
df_fill = df_elements[df_null_sum==1]
df_fill = df_fill.apply(lambda x: x.fillna(fill_values))
```

```python
df_fill_all = pd.concat([df_fill, df_elements[df_null_sum!=1]], axis=0).sort_index()
df_new = pd.concat([df_fill_all, df_properties], axis=1)
```

```python
#drop duplications
df_no_duplicate= df_new.drop_duplicates(subset =df_new.columns.tolist()[:12])
```

```python
property_names= df_missing.columns.tolist()[12:]
df_list = []
num_missing_threthod=3
for i in range(6):
    #drop missing value in property
    df = df_no_duplicate.iloc[:, list(range(0,12)) + [12+i]].dropna(subset=[property_names[i]])
    df_null = df.isnull().sum(axis=1)
    #keep the rows with less than 3 missing values and fill the rest of missing values with -1
    df_fill = df[df_null<num_missing_threthod].fillna(-1)
    df_list.append(df_fill)
    df_fill.to_csv('preprocessed/data_fillna_property'+str(i+1)+'.csv', index=False)
    train_, test_ = train_test_split(df_fill, test_size=0.3, random_state=4)
    train_.to_csv('preprocessed/p20_train_fillna_property'+str(i+1)+'.csv', index=False)
    test_.to_csv('preprocessed/p20_test_fillna_property'+str(i+1)+'.csv', index=False)
```
