```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
toy_alloy_data = pd.read_csv("toy_alloy_data.csv")
X_name = [f'element {i}' for i in range(1,13)]
Y_name = [f'property {i}' for i in range(1,7)]
```

```python
toy_alloy_data.describe()
```

```python
fea_data = toy_alloy_data.iloc[:,:12]
fea_data = fea_data.drop_duplicates()
fea_data.shape
```

```python
group_fea = toy_alloy_data.groupby(X_name)
```

```python
diff_label = set()
for key, value in group_fea:
    #print(key)
    for name in Y_name:
        if value[name].value_counts().shape[0] > 1:
            diff_label.add(name)
diff_label
```

## feature都相同的情况下，只出现了property6不相同的case
## 删除所有特征重复的项 property6取出现频次最高的一个数字

```python
remain_index = []
for key, value in group_fea:
    max_freq_value = value['property 6'].value_counts().index[0]
    use_idx = value[value['property 6'] == max_freq_value].index[0]
    remain_index.append(use_idx)
```

```python
len(remain_index)
```

```python
new_df = toy_alloy_data.iloc[remain_index].reset_index(drop=True)
new_df
```

```python
new_df.to_csv('toy_alloy_data_clean.csv', index=None)
```
